```python
normal = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]
```

## Undersampling ##

## Building a dataset containing equal distribution of normal and  fraudulent Transactions ##

```python
## fraud transactions is 492
normal_sample = normal.sample(n=492)
```

## concatenating the two dataframes ##

```python
new_data = pd.concat([normal_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

### shuffling the data ###

```python
new_data = new_data.sample(frac=1,random_state=42).reset_index(drop=True)
```

```python
X = new_data.drop(columns = 'Class', axis=1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Splitting the data into training and test data ##

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y,random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training ##

```python
model = LogisticRegression()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
## training the logistic Regression model with training data ##
model.fit(X_train, Y_train)
```

## Model Evaluation ##

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Testing data : ', test_data_accuracy)
```

## The model performed well with an accuracy score of 94% ##
